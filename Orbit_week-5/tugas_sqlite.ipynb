```python
import sqlite3
import pandas as pd
```

```python
connection = sqlite3.connect('DB/data.sqlite')
result = connection.cursor()
```


1. Tampilkan data customer yg mempunyai opening amount diatas 7000.

```python
query = "SELECT * FROM customer WHERE OPENING_AMT >= '7000';"
result.execute(query)
results = result.fetchall()
cols = [column[0] for column in result.description]
data = pd.DataFrame(results, columns= cols)
data
```

2. Tampilkan informasi order yg dilakukan melalui agensi dengan code A003.

```python
query = "SELECT * FROM orders WHERE AGENT_CODE = 'A003';"
result.execute(query)
results = result.fetchall()
cols = [col[0] for col in result.description]
data = pd.DataFrame(results, columns= cols)
data
```

3. Tampilkan data customer yg tinggal di New York, Bangalore, Mumbai, dan Torento.




```python
query = "SELECT * FROM customer WHERE WORKING_AREA IN ('New York', 'Bangalore', 'Mumbai')  "
result.execute(query)
results = result.fetchall()
cols = [col[0] for col in result.description]
data = pd.DataFrame(results, columns=cols)
data
```

4. Tampilkan informasi customer yang mempunyai payment amount diatas 7000 dan urutkan berdasarkan opening amountnya dari yg terbesar


```python
query = "SELECT * FROM customer WHERE PAYMENT_AMT >= '7000' ORDER BY OPENING_AMT ASC"
result.execute(query)
results = result.fetchall()
cols = [col[0] for col in result.description]
data = pd.DataFrame(results, columns=cols)
data
```


5. Tampilkan data agensi yang tidak bekerja di london dan memiliki comission lebih dari 0.12

```python
query = "SELECT * FROM agents WHERE WORKING_AREA != 'London' AND COMMISSION >= '0.12'"
result.execute(query)
results = result.fetchall()
cols = [col[0] for col in result.description]
data = pd.DataFrame(results, columns=cols)
data
```
