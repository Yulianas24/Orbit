```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np 
import os
```

```python
circles = cv2.imread('file/circle.png')
new = cv2.cvtColor(circles, cv2.COLOR_BGR2RGB)
plt.imshow(new)
```

```python
row =900
col =1500
cropped_image = new[row:row+1400,col:col+1400]
plt.imshow(cropped_image)
```

```python
new_rotate = cv2.rotate(new, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(new_rotate)
```

```python
width,height,dimension = new.shape
new_resize = cv2.resize(new, (height//20,width//20))
plt.imshow(new_resize)
```
