```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Mengumpulkan data dari file CSV

```python
cust = pd.read_csv('file\Customer.csv')
prod = pd.read_csv('file\prod_cat_info.csv')
tran = pd.read_csv('file\Transactions.csv')
```

## 1. Menghitung Jumlah baris data yang tersedia
* Dengan menggunakan fungsi len pada cust, prod, dan tran

```python
print("Jumlah Customer :",len(cust))
print("Jumlah Produk :",len(prod))
print("Jumlah Transaksi :",len(tran))
```

### Melakukan Merge antara tabel cust dan tran, sesuai dengan 'customer_id' dan 'cust_id'

```python
cust_tran = pd.merge(cust,tran,left_on='customer_Id',right_on='cust_id')
cust_tran
```

### 2. Mendeskripsikan statistika data sederhana (mean, max, min,..) menggunakan fungsi pandas
* menghitung fungsi mean,max, dan min pada total_amt menggunakan fungsi agg, yang dikelompokan sesuai dengan city_code menggunakan fungsi groupby


```python
result = cust_tran.groupby('city_code').agg({'total_amt': ['mean', 'max', 'min']})
result
```

### 3. menentukan berapa banyak pelanggan berdasarkan jenis kelamin (men vs women)


```python
gender = cust_tran.groupby(by=["Gender"]).count()
gender['transaction_id'].plot.bar(color=['red', 'blue'])
gender[['transaction_id']]
```

Dari data diatas diperoleh data pelanggan pria = 11811, dan pelanggan perempuan = 11233

### Menggabungkan semua data

```python
allData = pd.merge(cust_tran,prod,left_on=['prod_cat_code','prod_subcat_code'],right_on=['prod_cat_code','prod_sub_cat_code'])
allData
```

### 4. Menentukan product kategori (prod_cat) mana saja yang paling banyak diminati laki-laki 

```python
maleOnly = allData.drop(allData[ allData['Gender'] == 'F' ].index)
maleOnly = maleOnly.dropna()

m_result = maleOnly.groupby('prod_cat').agg({'prod_cat_code': ['count']})
m_result.plot(kind='bar', figsize = (8,5),width = 0.8)
```

Dari data diatas, dapat disimpulkan bahwa product kategori yang paling diminati pria yaitu mulai dari books, electronics, home and kitchen, clothing, footwear, yang terakhir bags

### 5. Menentukan product kategori (prod_cat) mana saja yang paling banyak diminati perempuan.

```python
femaleOnly = allData.drop(allData[ allData['Gender'] == 'M' ].index)
femaleOnly = femaleOnly.dropna()

f_result = femaleOnly.groupby('prod_cat').agg({'prod_cat_code': ['count']})
f_result.plot(kind='bar', figsize = (8,5),width = 0.8)
```

Dari data diatas, dapat disimpulkan bahwa product kategori yang paling diminati pria yaitu mulai dari books, electronics, home and kitchen, clothing, footwear, yang terakhir bags

### 6. Mana y store type yang paling banyak diakses oleh pelanggan untuk mencari produk di perusahaan retail ini

```python
store = allData.groupby(by=["Store_type"]).count()
store['cust_id'].plot.bar(figsize = (8,5),width = 0.8)
```

Dari data diatas dapat diambil kesimpulan bahwa store type yang paling banyak digunakan yaitu e-shop

### 7. Mana y bulan dengan penjualan tertinggi dan terendah selama satu tahun.

```python
# Menambahkan kolom bulan dan tahun kedalam tabel

allData['month'] = pd.DatetimeIndex(allData['tran_date']).month
allData['year'] = pd.DatetimeIndex(allData['tran_date']).year

#menjumlahkan data dengan grub by multi index month dan year
penjualan = allData.groupby(by = ['month','year']).sum('total_amt')
# Menampilkan grafik batang untuk nilai total_amt dengan indeks mulan dan tahun
penjualan['total_amt'].unstack().plot(kind = 'bar', figsize = (15,5),width = 0.8)
```

* Dari data diatas terlihat bahwa pada tahun 2011 penjualan tertinggi ada di bulan September, dan penjualan terendah ada di bulan Januari
* Pada tahun 2012 penjualan tertinggi berada di bulan Desember, dan penjualan terendah ada di bulan Februari
* Pada tahun 2013 penjualan tertinggi ada di bulan Oktober, dan penjualan terendah ada di bulan Februari
* Pada tahun 2014 penjualan tertinggi berada di bulan Januari, dan penjualan terendah ad di bulan Mei
